```python
import pandas as pd

# Cargar los DataFrames desde los archivos CSV previamente guardados
ruta_games = '/content/drive/My Drive/PI MLOps - STEAM/Games.csv'
ruta_reviews = '/content/drive/My Drive/PI MLOps - STEAM/Review.csv'
ruta_items = '/content/drive/My Drive/PI MLOps - STEAM/Items.csv'
```

```python
#Traigo el primer dataFrame
df_games = pd.read_csv(ruta_games)
```

```python
eliminar_columnas = ['user_id' ]
df_games = df_games.drop(eliminar_columnas, axis=1)
```

```python
eliminar_columnas = ['item_name','playtime_forever']
df_games = df_games.drop(eliminar_columnas, axis=1)
```

```python
df_reviews = pd.read_csv(ruta_reviews)
```

```python
df_reviews
```

```python
games_review = pd.merge(df_games, df_reviews, on='item_id')
```

```python
games_review
```

```python
eliminar_columnas = ['item_id']
games_review = games_review.drop(eliminar_columnas, axis=1)
```

```python
df = games_review.drop_duplicates(subset=['user_id'])
```

```python
df_items = pd.read_csv(ruta_items)
```

```python
total = pd.merge(df, df_items, on='user_id')
```

```python
total['release_date'] = total['release_date'].str.extract(r'(\d{4})')
```

```python
total['genres'] = total['genres'].apply(lambda x: x.replace("'", "").strip("[]"))
```

```python
ruta_guardar = '/content/drive/My Drive/PI MLOps - STEAM/Total.csv'
total.to_csv(ruta_guardar, index=False)  # index=False evita que se guarde el índice del DataFrame en el archivo
```
